```python
#RODAR ESSA CÉLULA UMA VEZ POR AMOSTRA

import juncao
import numpy as np

Res=[]
my_instrument = juncao.teste_equipamento()
resolucao= juncao.resolucao(primeiro_valor=0.005,ultimo_valor=0.05,passo=0.001)
```

```python


idut,Vdut = juncao.medicao(my_instrument,resolucao) 
Res=juncao.grafico(idut,Vdut,Res)
np.savetxt('valorees_resistencias.txt', Res, newline='\n')

```

```python
import matplotlib as plt
valores=np.loadtxt('juncoes_10k.txt') 
juncoes=[]
j=0
curto=0
total = len(valores)
aberto = 0
for i in valores:
    if i < 1000:
        valores[j]=0
        curto=curto+1
    if(i<100000 and i>1000):
        juncoes.append(i)
    if i > 100000:
        valores[j]=1
        aberto = aberto+1
    j=j+1

print(f"sao "+str(curto/len(valores)*100)+"% das junções em curto")
print(f"sao "+str(aberto/len(valores)*100)+"% das junções em aberto")
print(f"sao "+str((total-curto-aberto)/len(valores)*100)+"% de junções boas")

plt.hist(juncoes, 5, rwidth=0.9)
plt.show()

```
